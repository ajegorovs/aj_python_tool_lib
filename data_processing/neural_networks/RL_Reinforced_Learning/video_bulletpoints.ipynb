Stochastic vs deterministic MDP
1.  Deterministic
    1. Rewards are constant
    1. Actions bring you to to specific target state
1.  Stochastic
    1. Rewards are sampled from distribution
    1. Performing an action may bring you to a semi-random target state
    
1.  $p(s,a,s^\prime,r)$
    1.  causality order
    1.  full information, $P_{s,a,s^\prime,r}$ matrix, chain
        $$p(s,a,s^\prime,r) = p(s) \cdot p(a|s) \cdot p(s^\prime|s,a) \cdot p(r|s,a,s^\prime)$$
    1.  normalization, environment vs agent controlled
    1.  condition on $s, a$
        $$p(s,a,s^\prime,r|s,a) = \frac{p((s \cap a \cap s^\prime \cap r) \cap (s \cap a))}{p(s \cap a)} = \frac{p(s \cap a \cap s^\prime \cap r)}{p(s \cap a)}= $$
        $$=\frac{p(s) \cdot p(a|s) \cdot p(s^\prime|s,a) \cdot p(r|s,a,s^\prime)}{p(s, a)}=\frac{p(s) \cdot p(a|s) \cdot p(s^\prime|s,a) \cdot p(r|s,a,s^\prime)}{p(s)\cdot p(a|s)} \Rightarrow$$
        $$\boxed{p(s^\prime,r|s,a) = p(s^\prime|s,a) \cdot p(r|s,a,s^\prime)}$$
        
1.  setup:
    1. actions - left & right
    1. taking action results in various state transitions   
    1. rewards for different 'leaps'
    1. fill matrix with number of events 
    1. example with number of rewards
1.  expected reward for <i>state-action-next-state</i> triplet
    1. $$p(r|s,a,s^\prime) = \frac{p(s^\prime,r|s,a)}{p(s^\prime|s,a)}$$
    1. $$r(s,a,s^\prime) = \mathbb{E}_r[p(r|s,a,s^\prime) ] = \sum_r  r\cdot p(r|s,a,s^\prime) = \frac{\sum_r  r\cdot p(s^\prime,r|s,a)}{p(s^\prime|s,a)}$$
    1. show example

1.  state-action value $\mathbb{E}_{s^\prime}[r(s,a,s^\prime)]$
    1.  elements of derivation
        $$r(s,a) =  \mathbb{E}_{s^\prime}[r(s,a,s^\prime)] = \sum_{r \in R} \sum_{s^\prime \in S} r \cdot p(s^\prime, r|s,a)$$
    1.  show example

1.  Conclusions
    * in stochastic MDP rewards are random and result of actions is random
    * probability $p(s^\prime,r|s,a)$ describes dynamics of a environment, not agent
