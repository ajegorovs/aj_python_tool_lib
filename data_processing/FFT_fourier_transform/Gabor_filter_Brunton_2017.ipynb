Gabor transform.

It is a Short Time Fourier Transform (STFT) that enables to analyze transient (non periodic) signals.



Quadratic chirp. Copied from book's worked examples because there is nothing to do. We can apply Gabor filter to cosine signal where frequency changes from f1 to f2 during time span t1-t0

```python

from matplotlib import cm
cmap = cm.PRGn
dt = 0.001
t = np.arange(0,2,dt)
f0 = 50
f1 = 250
t1 = 2
freqs = t*(f0 + (f1-f0)*np.power(t,2)/(3*t1**2))
x = np.cos(2*np.pi*freqs)
fig, _axs = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)
#fig.subplots_adjust(hspace=0.3)
axs = _axs.flatten()

p1 = axs[0].plot(t, freqs)
p2 = axs[1].specgram(x, NFFT=128, Fs=1/dt, noverlap=120,cmap=cmap)
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Frequency (Hz)')

```
