```python
import cv2
import matplotlib.pyplot as plt
import numpy as np, os

BGR = cv2.imread(os.path.join('data_processing','media','hsb_colourspace3-1.png'), flags=cv2.IMREAD_COLOR)
hsv = cv2.cvtColor(BGR, cv2.COLOR_BGR2HSV)
RGB = cv2.cvtColor(BGR, cv2.COLOR_BGR2RGB)

fig, ax = plt.subplots(2,2, figsize = (15,7))
ax = ax.flatten()
ax[0].imshow(RGB)
ax[0].axis('off')
ax[0].set_title('RGB')
for channel in range(3):
    im = ax[channel+1].imshow(hsv[:,:,channel], 'gray')
    ax[channel+1].axis('off')
    ax[channel+1].set_title(f'channel: {channel}')
fg_color = 'black'
fig.colorbar(im, ax=ax.ravel().tolist())
#cb = plt.colorbar(im,orientation='horizontal')
# cb.set_label('colorbar label', color=fg_color)
# cb.outline.set_edgecolor(fg_color)
# cb.ax.yaxis.set_tick_params(color=fg_color)
#plt.setp(plt.getp(cb.ax.axes, 'yticklabels'), color=fg_color)
```
