# Symbolic stuff with sumpy.

```python
from sympy import *
```

```python
x1,x2, c1, c2  = symbols("x_1, x_2, c_1, c_2")

q11, q12, q21, q22  = symbols("Q_{11} Q_{12} Q_{21} Q_{22}")

Q = Matrix([[q11,q12],[q21,q22]])
x = Matrix([[x1,x2]]).T
c = Matrix([[c1,c2]]).T

f = 1/2*(x.T @ Q @ x) +c.T @ x
inp = collect(f.expand()[0], x1*x2)
display(inp)
```

```python
from sympy.abc import x,y,z
import matplotlib.pyplot as plt
sl = solve(6*x*x+4*y-24   , y, dict=False)[0]

display(Eq(y,sl))

f1 = lambdify([x,y,z],sl) # y,z dont do anything

xses = range(7)
yses = [f1(i,0,0) for i in xses]
plt.plot(xses, yses, 'o')
```
